```python
# Let's iterate over numbers 1, 2, ... 10
# In python we start counting from 0, and the range function is not inclusive
# Initialize
sum_of_numbers = 0
for i in range(1, 11):
    sum_of_numbers = sum_of_numbers + i
    
sum_of_numbers
```

We prototyped... let's define a function.

```python
def add_numbers_up_to(end : int) -> int:
    """
    This function sums numbers from 1 up until an end number
    
    Parameters
    ----------
    end (int) how many numbers we want to add
    
    Returns
    -------
    sum_of_numbers (int) sum from 1 until end
    
    """
    
    sum_of_numbers = 0
    for i in range(1, end):
        sum_of_numbers = sum_of_numbers + i
    
    return sum_of_numbers
```
